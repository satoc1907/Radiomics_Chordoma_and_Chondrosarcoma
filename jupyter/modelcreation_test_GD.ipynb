```python
def make_df_dT2_txt (a, b, c, d, e):
    
    a.index = a ['Dnum']
    b.index = b ['Dnum']
    c.index = c ['Dnum']
    d.index = d ['Dnum']
    e.index = e ['Dnum']

    #Data extraction
    a = a.iloc [:, 2: 101]
    b = b. iloc [:, 2: 119]
    c = c.iloc [:, 2: 119]
    d = d. iloc [:, 2: 146]
    e = e.iloc [:, 2: 47]
   

    # Integration of data
    d_all = pd.concat ([a, b, c, d, e], axis = 1)

    # Column name change
    d_all = d_all.rename (columns = lambda s: "dT2" + s)
    
    return d_all
```

```python
def make_df_dGD_txt (a, b, c, d, e):
    
    a.index = a ['Dnum']
    b.index = b ['Dnum']
    c.index = c ['Dnum']
    d.index = d ['Dnum']
    e.index = e ['Dnum']

    #Data extraction
    a = a.iloc [:, 2: 101]
    b = b. iloc [:, 2: 119]
    c = c.iloc [:, 2: 119]
    d = d. iloc [:, 2: 146]
    e = e.iloc [:, 2: 47]
   

    # Integration of data
    d_all = pd.concat ([a, b, c, d, e], axis = 1)

    # Column name change
    d_all = d_all.rename (columns = lambda s: "dGD" + s)
    
    return d_all
```
